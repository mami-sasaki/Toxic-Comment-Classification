# Toxic Comment Classification

As a NLP Engineer, I've always been interested in working on maulicious comments such as cyberbullying, abuse, hate speech, violent threats to create safe and healthy cyperspace and to protect kids from tragic outcomes.  

In this work, I built a binary classifcation model which is capable of detecting toxic comments. 
The data are downloaded from toxic comment classificaiton challenge dataset from Kaggle which contains omments from Wikipedia’s talk page edits. The training data contains 159571 and test contains 153164 comments and they are classified as either toxic (class = 1) or non toxic (class = 0).

I used this training dataset to train a model that can take in new comments in test dataset and predict whether they are toxic or not.
The data also include different types of toxicity which are toxic, severe toxic, threats, obscenity, insults, and identity-based hate. I would like to further investigate and build multiclassificaiton model as a next step in the near future. 

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer 
from sklearn.naive_bayes import MultinomialNB

df_train = pd.read_csv('train.csv')
df_train.head(15)
```

I was curisous how bad the comments look like when all the categories are classified as 1 so here it is.

```python
bad_bad_comment = 'toxic == 1 & severe_toxic ==1 & obscene == 1 & threat == 1 & insult == 1 & identity_hate == 1' 
bad_comments = df_train.query(bad_bad_comment)
bad_comments
```

Also how big the proportion of toxic comments in a whole dataset? 

```python
toxic = df_train[df_train['toxic'] == 1]
len(toxic)/len(df_train) 
```

9.6% of comments are toxic! It's almost 10% of data :( 

And I also wanted to see the size in the differnt combination of toxicity. 

```python
count = df_train.groupby(['toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']).size() 
count
```

First step is to create a word vector using BOW (bag of words) with stopwords filter.  

```python
word_bagger = CountVectorizer(stop_words="english")
word_bagger.fit(df_train['comment_text'])
```

```python
total_words = word_bagger.get_feature_names()
len(total_words)
```

For this work, I decided to use Multinomial Naive Bayes model as it is suitable for data where the features are counts. 

```python
y = df_train['toxic']
X = word_bagger.transform(df_train['comment_text'])
toxic_nb = MultinomialNB()
toxic_nb.fit(X, y)
```

```python
toxic_nb.score(X,y)
```

95.5% training accuracy is predicted.
Now apply test data to the trained model. 

```python
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
test_bagger = word_bagger.transform(df_test['comment_text'])
test_bagger.shape
```

```python
df_test['predicted'] = toxic_nb.predict(test_bagger)
df_test['probability'] = toxic_nb.predict_proba(test_bagger)[:,1]
df_test
```

Let's see if it works with my own constructed data.

```python
c1 = "I think coronavirus makes people nuts."
c2 = "ppl are fucking selfish pieces of shit."
c3 = "my dog is a bitch." # true statement
c4 = "japanese ppl don't shower." # sarcasm! 
c5 = "I think you are very ugly." # why not toxic? 
test_bagger2 = word_bagger.transform([c1,c2,c3,c4,c5])
```

```python
toxic_nb.predict(test_bagger2)
```

```python
toxic_nb.predict_proba(test_bagger2)
```

```python
toxic_nb.predict_proba(test_bagger2)[:,1]
```

It looks pretty good overall but there's definitely some room for improvement. For instance, c5 can be read very ambigious and c4 is a sarcasm and not necessarily toxic which is known as one of the hardest chalelnges NLP appliaction is dealing with.

### Evaluation

Toxic comment classification challenge data from Kaggle also contains a dataset for evaluation conveniently so I compared my model's prediction against the evalution dataset.

```python
df_test_label = pd.read_csv('test_labels.csv')
df_test_label
```

-1 means these comments are not used for evaluation so I only extracted data where 'toxic' value is 0 (=non toxic) or 1 (=toxic) and added a column 'matched' to compare my model prediciton and manually evaluated result.  

```python
df_evaluated = df_test_label[df_test_label['toxic'] > -1]
df_final = pd.merge(df_evaluated, df_test, how='left', on='id')
df_final['matched'] = np.where(df_final['toxic'] == df_final['predicted'], 'True', 'False')  
df_final
```

```python
matched = df_final.matched.value_counts().loc['True']
matched/len(df_final)
```

Overall accuracy is 91.8%! 
To summarize, I present performance measures of confusion matrix, precision, recall and F1 score. 

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report 

actual = df_final['toxic']
predicted = df_final['predicted']
results = confusion_matrix(actual,predicted)
TP = results[1,1]
TN = results[0,0]
FP = results[0,1]
FN = results[1,0]
print ('Confusion Matrix :')
print(results) 
print('[[TN ','FP','\n  FN',' TP]]')
print ('\nAccuracy Score :',accuracy_score(actual, predicted)) 
print ('\nReport : ')
print (classification_report(actual, predicted)) 
```

<b>Classification Rate/Accuracy:</b><br>
Accuracy = (TP + TN) / (TP + TN + FP + FN)<br>
         = (4078 + 54656) /(4078 + 54656 + 3232 + 2012)<br> 
         = 0.9180343242989778

<b>Recall</b>: <i>Recall gives us an idea about when it’s actually yes, how often does it predict yes.</i><br>
Recall = TP / (TP + FN)<br>
       = 4078 / (4078 + 2012)<br> 
       = 0.6696223316912973

<b>Precision</b>: <i>Precsion tells us about when it predicts yes, how often is it correct.</i><br>
Precision = TP / (TP + FP)<br>
          = 4078 / (4078 + 3232)<br>
          = 0.5578659370725034

<b>F1-score:</b> <i>The weighted harmonic mean of the test's precision and recall</i><br>
F1-score = (2 * Recall * Precision) / (Recall + Presision)<br>
         = (2 * 0.67 * 0.56) / (0.67 + 0.56)<br> 
         = 0.6100813008130082
